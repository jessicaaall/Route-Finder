### **2.1. Rancangan Simulasi**

Rancangan simulasi pencarian rute berikut merupakan bentuk sederhana dari pencarian rute, yang menggunakan maze sebagai peta sehingga pencarian rute tidak melakukan input nama jalan, melainkan titik koordinat dari maze tersebut. Simulasi sistem secara khusus hanya menerapkan pencarian rute terpendek dan menghitung jarak rute tersebut dalam bentuk langkah atau step.

1. PROSES : Peta yang telah disediakan diproses oleh sistem sehingga sistem mengetahui elemen dari setiap titik-titik koordinat pada peta.

2. OUTPUT : Sistem akan menampilkan peta berupa maze terlebih dahulu kepada user sehingga user dapat melihat titik-titik koordinat jalan pada peta.

3. OUTPUT : Sistem menampilkan daftar titik-titik koordinat jalan kepada user untuk memudahkan user mengetahui titik-titik koordinat jalan pada peta.

4. INPUT : User melakukan input titik koordinat asal dan titik koordinat tujuan yang ingin ditujunya.

5. PROSES : Sistem melakukan pengecekan apakah titik koordinat asal dan titik koordinat tujuan yang diinput oleh user sesuai dengan titik-titik koordinat jalan yang ada pada peta. Jika sesuai, titik koordinat asal dan titik koordinat tujuan diterima sistem untuk diproses lebih lanjut.

6. OUTPUT : Jika titik koordinat asal dan/atau titik koordinat tujuan yang diinput oleh user tidak sesuai dengan titik-titik koordinat jakan yang ada pada peta, maka sistem akan menampilkan output yang memberitahu user bahwa titik koordinat asal dan/atau titik koordinat tujuan yang diinput salah.

7. PROSES : Elemen dari titik koordinat asal yang diinput user akan dihilangkan dan diisi dengan suatu simbol yang akan menjadi penanda rute yang dilalui. Titik koordinat selanjutnya tersebut diperoleh dengan melakukan 1 langkah ke kanan, 1 langkah ke kiri, 1 langkah ke atas, atau 1 langkah ke bawah dari titik koordinat asal, dengan melakukan pengecekan apakah 1 langkah ke kanan, 1 langkah ke kiri, 1 langkah ke atas, atau 1 langkah ke bawah dari titik koordinat asal tersebut merupakan jalan yang dapat dilalui. Pada titik koordinat selanjutnya tersebut, juga dilakukan penghilangan elemen dan diisi dengan simbol penanda rute yang dilalui. Titik koordinat selanjutnya terus dicari serta dilakukan penghilangan elemen dan pengisian simbol penanda rute, hingga titik koordinat selanjutnya mencapai titik koordinat tujuan.

8. PROSES : Sistem menghitung jumlah bertambahnya simbol penanda rute yang dilalui dari titik koordinat asal untuk memperoleh banyak step dari titik koordinat asal ke titik koordinat tujuan.

9. OUTPUT : Sistem menampilkan peta yang telah diisi dengan simbol penanda rute yang dilalui dari titik koordinat asal hingga mencapai titik koordinat akhir.

10. OUTPUT : Sistem menampilkan jumlah bertambahnya simbol penanda dari rute yang dilalui dari titik koordinat asal hingga mencapai titik koordinat akhir, sebagai banyak step yang ditempuh.




###**2.2. Algoritma Pseudocode**

```python
matrikspeta = [list('HHHHHHHHHHHHHHHHHHHHHHHHHHHHHH'),    {gambar peta}
               list('H   HHHH   HHHH    HH   HHH  H'),
               list('HH HH    HHHH   HHHH  HHHH  HH'),
               list('HH    HH  HH   HHHHH   HHH   H'),
               list('HHH HHHHH HHHH  HH   HH    HHH'),
               list('H   HHHH    HHH HHHH    HH  HH'),
               list('HHH HH   HH  H   HHH  HHH  HHH'),
               list('HH   HHH HHH  HH  H  HHH  HHHH'),
               list('H  HHH   HH  HHHH    HH    HHH'),
               list('HH  HH   HH  H    HH  HHH  HHH'),
               list('H   HHHH     HH HHHHH  HHH   H'),
               list('HHH H    HH      HHH  HHH  HHH'),
               list('HH    H HHHH  HH  H  HHHHH   H'),
               list('H  HHHH   HH  H   HH   HHH  HH'),
               list('HHHHHHHHHHHHHHHHHHHHHHHHHHHHHH'),]


output("------- ILUSTRASI PETA -------")
i traversal [1..panjang baris matrikspeta]
  j traversal [1..panjang kolom matrikspeta]
    output(matrikspeta[i][j], end='')        {mencetak gambar peta untuk ditampilkan kepada user}
  output()

output()
output("NOTE : PADA IUSTRASI PETA, "H" PADA UJUNG KIRI PALING ATAS DIANGGAP SEBAGAI TITIK KOORDINAT (0,0) DENGAN TITIK KOORDINAT KE BAWAH DAN KE KANAN BERNILAI POSITIF")
output()
output()

daftarkoordinat = []
output("===== Daftar titik koordinat [x,y] yang dapat menjadi koordinat asal ataupun koordinat akhir =====")
j traversal [1..panjang kolom matrikspeta]
  i traversal [1..panjang baris matrikspeta]
    if (matrikspeta[i][j] == ' ') then          {jika matrikspeta baris ke-i kolom ke-j kosong}
      titikkoordinat <- [j,i]                   {titik koordinat adalah kolom ke-j, baris ke-1 (j sebagai titik x dan i sebagai titik y)}
      output(titikkoordinat, end=' ')           {mencetak titik-titik koordinat yang kosong pada peta yang dapat menjadi koordinat asal ataupun koordinat akhir user}
      daftarkoordinat.append(titikkoordinat)    {mengisi titik-titik koordinat tersebut ke daftar koordinat}
  output()

output()
output("===== Masukkan koordinat asal dan koordinat tujuan Anda berdasarkan daftar titik koordinat (x,y) yang telah ditampilkan =====")
output()

output("---- Silahkan isi koordinat asal Anda ----")
input(xasal)     {input titik koordinat x asal}
input(yasal)     {input titik koordinat y asal}
output()

output("---- Silahkan isi koordinat tujuan Anda ----")
input(xtujuan)      {input titik koordinat x tujuan}
input(ytujuan)      {input titik koordinat y tujuan}
output()
output()

{Definisi Fungsi stepselanjutnya}
function stepselanjutnya[(x, y)]
  {Membuat daftar step selanjutnya yang dapat dilakukan}
	return [([[x-1, y],     {step ke kiri sebanyak 1 satuan}
	          [x+1, y],     {step ke kanan sebanyak 1 satuan}
					  [x, y-1],     {step ke atas sebanyak 1 satuan (berdasarkan matriks, ke bawah dianggap bernilai positif)}
					  [x, y+1]])]   {step ke bawah sebanyak 1 satuan (berdasarkan matriks, ke bawah dianggap bernilai positif)}


{Definisi Fungsi carijalurterpendek}
function carijalurterpendek[(matrikspeta, koordinatasal, koordinattujuan)]
  {Mencari jalur terpendek untuk mencapai koordinat tujuan dari koordinat asal}
  carijalur <- [[koordinatasal]]
	koordinatdilalui <- [koordinatasal]
	while (carijalur != []) do
		jalursaatini <- carijalur.pop(0)         {elemen indeks ke-0 dari carijalur dihilangkan}
		koordinatsaatini <- jalursaatini[-1]     {mengambil array terakhir dari matriks jalursaatini}
		Xsaatini, Ysaatini <- koordinatsaatini
		if (koordinatsaatini == koordinattujuan) then                         {jika koordinatsaatini sudah sama dengan koordinattujuan}
			return [(jalursaatini)]                                             {mengembalikan elemen jalursaatini}
		for koordinatselanjutnya in stepselanjutnya(Xsaatini, Ysaatini)       {untuk koordinatselanjutnya yang ada di daftar step selanjutnya yang dapat dilakukan}
			Xselanjutnya, Yselanjutnya <- koordinatselanjutnya
			if (koordinatselanjutnya in koordinatdilalui) then                  {jika koordinatselanjutnya ada di koordinatdilalui}
				continue
			if (matrikspeta[Yselanjutnya][Xselanjutnya] == 'H') then            {jika pada matrikspeta baris ke-Yselanjutnya kolom ke-Xselanjutnya berisi "H"}
				continue
			carijalur.append(jalursaatini + [koordinatselanjutnya])             {pada matriks carijalur, ditambahkan matriks dari jalursaatini tambah koordinatselanjutnya}
			koordinatdilalui <- koordinatdilalui + [koordinatselanjutnya]       {pada matriks koordinatdilalui ditambahkan koordinatselanjutnya}


{ALGORITMA PROGRAM UTAMA}

koordinatasal <- [xasal, yasal]
koordinattujuan <- [xtujuan, ytujuan]

sum <- -1    {sum diberi nilai -1 agar pada saat tanda "." untuk koordinat asal dimunculkan menjadi terhitung dari 0}

if ((koordinatasal in daftarkoordinat) and (koordinattujuan in daftarkoordinat)) then    {jika koordinatasal dan koordinattujuan ada di daftarkoordinat}
  jalurterpendek <- carijalurterpendek(matrikspeta, koordinatasal, koordinattujuan)
  output("===== Step untuk mencapai koordinat tujuan dari koordinat asal =====")
  output()
  for koordinat in jalurterpendek
    x, y <- koordinat
    matrikspeta[y][x] <- '.'   {mengisi matrikspeta baris ke-y kolom ke-x dengan "."}
    sum <- sum + 1             {menjumlahkan banyak "."}
    for row in matrikspeta
      output(''.join(row))     {membuat hasil print matrikspeta tergabung sehingga tidak terdapat tanda "[ ]", " ' ", dan "," pada hasil print}
    output()
  output("---------------------------------------------------------------------")
  output()
  output("Banyak step yang ditempuh untuk mencapai titik koordinat tujuan", koordinattujuan, "dari titik koordinat asal", koordinatasal, "=", sum, "satuan")
else {jika koordinatasal dan/atau koordinattujuan tidak ada di daftarkoordinat}
  output("---------------------------------------------------------------------")
  output()
  output("Masukkan koordinat asal dan/atau koordinat tujuan Anda tidak ditemukan.")
  output("Mohon periksa kembali masukkan koordinat asal dan tujuan Anda sesuai dengan daftar titik koordinat yang ditampilkan di atas.")
```

## **Bagian 3 PROGRAM**


```python
# Program PencarianRute
# Menampilkan rute pada suatu peta dari masukkan titik koordinat asal ke masukkan titik koordinat tujuan
# Menghitung banyak langkah yang ditempuh untuk mencapai titik koordinat tujuan dari titik koordinat asal

# KAMUS
# matrikspeta : matriks of str
# daftarkoordinat, jalurterpendek : matriks of int
# i, j : int (indeks)
# titikkoordinat, koordinatasal, koordinattujuan, koordinat : array of int
# xasal, yasal, xtujuan, ytujuan, x, y, sum : int (elemen)

# ALGORITMA

matrikspeta = [list('HHHHHHHHHHHHHHHHHHHHHHHHHHHHHH'),    # gambar peta
               list('H   HHHH   HHHH    HH   HHH  H'),    # gambar peta dapat Anda ubah
               list('HH HH    HHHH   HHHH  HHHH  HH'),
               list('HH    HH  HH   HHHHH   HHH   H'),
               list('HHH HHHHH HHHH  HH   HH    HHH'),
               list('H   HHHH    HHH HHHH    HH  HH'),
               list('HHH HH   HH  H   HHH  HHH  HHH'),
               list('HH   HHH HHH  HH  H  HHH  HHHH'),
               list('H  HHH   HH  HHHH    HH    HHH'),
               list('HH  HH   HH  H    HH  HHH  HHH'),
               list('H   HHHH     HH HHHHH  HHH   H'),
               list('HHH H    HH      HHH  HHH  HHH'),
               list('HH    H HHHH  HH  H  HHHHH   H'),
               list('H  HHHH   HH  H   HH   HHH  HH'),
               list('HHHHHHHHHHHHHHHHHHHHHHHHHHHHHH'),]


print('------- ILUSTRASI PETA -------')
for i in range (0,len(matrikspeta)) :
  for j in range (0,len(matrikspeta[0])) :
    print(matrikspeta[i][j], end='')        # mencetak gambar peta untuk ditampilkan kepada user
  print()

print()
print('NOTE : PADA IUSTRASI PETA, "H" PADA UJUNG KIRI PALING ATAS DIANGGAP SEBAGAI TITIK KOORDINAT (0,0) DENGAN TITIK KOORDINAT KE BAWAH DAN KE KANAN BERNILAI POSITIF')
print()
print()

daftarkoordinat = []
print('===== Daftar titik koordinat [x,y] yang dapat menjadi koordinat asal ataupun koordinat akhir =====')
for j in range (0,len(matrikspeta[0])):
  for i in range (0,len(matrikspeta)):
    if (matrikspeta[i][j] == ' '):              # jika matrikspeta baris ke-i kolom ke-j kosong
      titikkoordinat = [j,i]                    # titik koordinat adalah kolom ke-j, baris ke-1 (j sebagai titik x dan i sebagai titik y)
      print(titikkoordinat, end=' ')            # mencetak titik-titik koordinat yang kosong pada peta yang dapat menjadi koordinat asal ataupun koordinat akhir user
      daftarkoordinat.append(titikkoordinat)    # mengisi titik-titik koordinat tersebut ke daftar koordinat
  print()

print()
print('===== Masukkan koordinat asal dan koordinat tujuan Anda berdasarkan daftar titik koordinat (x,y) yang telah ditampilkan =====')
print()

print('---- Silahkan isi koordinat asal Anda ----')
xasal = int(input('Masukkan titik x asal Anda = '))     # input titik koordinat x asal
yasal = int(input('Masukkan titik y asal Anda = '))     # input titik koordinat y asal
print()

print('---- Silahkan isi koordinat tujuan Anda ----')
xtujuan = int(input('Masukkan titik x tujuan = '))      # input titik koordinat x tujuan
ytujuan = int(input('Masukkan titik y tujuan = '))      # input titik koordinat y tujuan
print()
print()

# Definisi Fungsi stepselanjutnya
def stepselanjutnya(x, y) :
  # Membuat daftar step selanjutnya yang dapat dilakukan

  # KAMUS LOKAL
  # tidak perlu variabel khusus

  # ALGORITMA
	return [[x-1, y],   # step ke kiri sebanyak 1 satuan
	        [x+1, y],   # step ke kanan sebanyak 1 satuan
					[x, y-1],   # step ke atas sebanyak 1 satuan (berdasarkan matriks, ke bawah dianggap bernilai positif)
					[x, y+1]]   # step ke bawah sebanyak 1 satuan (berdasarkan matriks, ke bawah dianggap bernilai positif)


# Definisi Fungsi carijalurterpendek
def carijalurterpendek(matrikspeta, koordinatasal, koordinattujuan) :
  # Mencari jalur terpendek untuk mencapai koordinat tujuan dari koordinat asal

  # KAMUS LOKAL
  # carijalur, koordinatdilalui, jalursaatini : matriks of int
  # koordinatsaatini, koordinatselanjutnya : array of int
  # Xsaatini, Ysaatini, Xselanjutnya, Yselanjutnya : int

  # ALGORITMA
	carijalur = [[koordinatasal]]
	koordinatdilalui = [koordinatasal]
	while (carijalur != []) :
		jalursaatini = carijalur.pop(0)         # elemen indeks ke-0 dari carijalur dihilangkan
		koordinatsaatini = jalursaatini[-1]     # mengambil array terakhir dari matriks jalursaatini
		Xsaatini, Ysaatini = koordinatsaatini
		if (koordinatsaatini == koordinattujuan) :                          # jika koordinatsaatini sudah sama dengan koordinattujuan
			return jalursaatini                                               # mengembalikan elemen jalursaatini
		for koordinatselanjutnya in stepselanjutnya(Xsaatini, Ysaatini) :   # untuk koordinatselanjutnya yang ada di daftar step selanjutnya yang dapat dilakukan
			Xselanjutnya, Yselanjutnya = koordinatselanjutnya
			if (koordinatselanjutnya in koordinatdilalui) :                   # jika koordinatselanjutnya ada di koordinatdilalui
				continue
			if (matrikspeta[Yselanjutnya][Xselanjutnya] == 'H') :             # jika pada matrikspeta baris ke-Yselanjutnya kolom ke-Xselanjutnya berisi "H"
				continue
			carijalur.append(jalursaatini + [koordinatselanjutnya])           # pada matriks carijalur, ditambahkan matriks dari jalursaatini tambah koordinatselanjutnya
			koordinatdilalui = koordinatdilalui + [koordinatselanjutnya]      # pada matriks koordinatdilalui ditambahkan koordinatselanjutnya


# ALGORITMA PROGRAM UTAMA

koordinatasal = [xasal, yasal]
koordinattujuan = [xtujuan, ytujuan]

sum = -1    # sum diberi nilai -1 agar pada saat tanda "." untuk koordinat asal dimunculkan menjadi terhitung dari 0

if ((koordinatasal in daftarkoordinat) and (koordinattujuan in daftarkoordinat)) :    # jika koordinatasal dan koordinattujuan ada di daftarkoordinat
  jalurterpendek = carijalurterpendek(matrikspeta, koordinatasal, koordinattujuan)
  print('===== Step untuk mencapai koordinat tujuan dari koordinat asal =====')
  print()
  for koordinat in jalurterpendek :
    x, y = koordinat
    matrikspeta[y][x] = '.'   # mengisi matrikspeta baris ke-y kolom ke-x dengan "."
    sum = sum + 1             # menjumlahkan banyak "."
    for row in matrikspeta :
      print(''.join(row))     # membuat hasil print matrikspeta tergabung sehingga tidak terdapat tanda "[ ]", " ' ", dan "," pada hasil print
    print()
  print('---------------------------------------------------------------------')
  print()
  print('Banyak step yang ditempuh untuk mencapai titik koordinat tujuan', koordinattujuan, 'dari titik koordinat asal', koordinatasal, '=', sum, 'satuan')
else :  # jika koordinatasal dan/atau koordinattujuan tidak ada di daftarkoordinat
  print('---------------------------------------------------------------------')
  print()
  print('Masukkan koordinat asal dan/atau koordinat tujuan Anda tidak ditemukan.')
  print('Mohon periksa kembali masukkan koordinat asal dan tujuan Anda sesuai dengan daftar titik koordinat yang ditampilkan di atas.')
```
